# Matplotlib Mastery

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

### Basic Line Plot 

```python
x = np.linspace(start=0, stop=10, num=100)
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

### Basic Scatter Plot 

```python
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
size = 1000 * np.random.rand(50)

plt.scatter(x=x, y=y, alpha=.5, s=size, c=colors)
plt.colorbar()
plt.show()
```

### Subplots 

```python
fig, axs = plt.subplots(2, 2, figsize=(8, 7))
axs[0, 0].plot(x, y)
axs[0, 0].set_title('Subplot 1')
axs[0, 1].scatter(x, y)
axs[0, 1].set_title('Subplot 2')
axs[1, 0].hist(y)
axs[1, 0].set_title('Subplot 3')
axs[1, 1].imshow(np.random.rand(10, 8))
axs[1, 1].set_title('Subplot 4')

plt.show()
```

### Bar Plot 

```python
labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 35, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]
x = np.arange(len(labels))
width = 0.35
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Men')
rects2 = ax.bar(x + width/2, women_means, width, label='Women')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
plt.show()
```

### Histogram 

```python
np.random.seed(42)
x = np.random.normal(size=1000)
plt.hist(x, bins=30)
plt.show()
```

### Pie Chart 

```python
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
explode = (0.1, 0, 0, 0)
fig, ax = plt.subplots()
ax.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True)
ax.axis('equal')
plt.show()
```

### Customizing Plot 

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, color='green', linestyle='dashed', linewidth=2, marker='o', markerfacecolor='blue', markersize=8)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.grid(True)
plt.show()

plt.savefig( 'plot.png', dpi=300, bbox_inches='tight')
```
